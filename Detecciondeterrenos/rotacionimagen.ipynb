```python
import cv2
import numpy as np
import glob 
def correct_orientation(img_rgb,dim,path_pattern:str="pattern.png"):
    img_rgb=cv2.imread(img_rgb)
    img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)
    template = cv2.imread(path_pattern,0)
    w, h = template.shape[::-1]
    image_ro=img_gray.copy()
    angulo=0
    an=[]
    le=0
    angulo_f=0
    for i in range(0,90,1):
        angulo=i
        M = cv2.getRotationMatrix2D((dim//2,dim//2), angulo, 1)
        image_ro = cv2.warpAffine(img_gray, M, (dim,dim))
        res = cv2.matchTemplate(image_ro,template,cv2.TM_CCOEFF_NORMED)
        threshold =.5
        loc = np.where( res >= threshold)
        com=len(loc[0])
        if com>0:
            an.append(angulo)
            if le<com:
                le=com
                angulo_f=angulo
                M = cv2.getRotationMatrix2D((dim//2,dim//2), angulo, 1)
                image_ro = cv2.warpAffine(img_rgb, M, (dim,dim))
                imagen_final=image_ro.copy()
    return angulo_f,imagen_final
```

```python
import tqdm
filenames=glob.glob(r"G:\Mi unidad\Equipo_Agua\Geo\Data\Naucalpan\Fragmentos\Hector/*.png")
for file in tqdm.tqdm(filenames):
    angulo,imagen=correct_orientation(file,1024, r"C:\Users\52551\Downloads\pattern.png")
    name = file.split('\\')[-1]
    # cv2.imshow("g",imagen)
    # cv2.waitKey()
    # cv2.destroyAllWindows()
    cv2.imwrite(r"G:\Mi unidad\Equipo_Agua\Geo\Data\Naucalpan\Fragmentos\Hector\Rotados/"+name,imagen)
```
