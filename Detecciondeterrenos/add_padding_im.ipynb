```python
import cv2
import numpy as np
import glob 
import pandas as pd
import os
import tqdm
```

```python
filenames=glob.glob(r"C:\Users\mfpen\Pictures\Clasificador\validacion/*/*.png")+glob.glob(r"C:\Users\mfpen\Pictures\Clasificador\validacion/*/*.PNG")
dirs=[]
filename=[]
for file in filenames:
    path=file.replace("\\","/").split("/")
    path[-3]="validacion_pad"
    dirs.append("/".join(path[:-1]))
    filename.append(path[-1])
imagenes=pd.DataFrame({"path":filenames,"dir":dirs,"filename":filename})
imagenes
```

```python
def add_padding(imagenes):
    for carpeta in imagenes["dir"].unique():
        try:
            os.mkdir(carpeta)
            
            for i in tqdm.tqdm(range(len(imagenes))):
                img = cv2.imread(imagenes["path"][i])
                old_image_height, old_image_width, channels = img.shape

                # create new image of desired size and color (blue) for padding
                new_image_width = 224
                new_image_height = 224
                color = (0,0,0)
                result = np.full((new_image_height,new_image_width, channels), color, dtype=np.uint8)

                # compute center offset
                x_center = (new_image_width - old_image_width) // 2
                y_center = (new_image_height - old_image_height) // 2

                # copy img image into center of result image
                try:
                    result[y_center:y_center+old_image_height, 
                        x_center:x_center+old_image_width] = img
                except:
                    continue

                cv2.imwrite(imagenes["dir"][i]+"/"+imagenes["filename"][i], result)
        except:
            pass   
    
    return 
```

```python
add_padding(imagenes)
```
