```python
import cv2
import numpy as np
import glob 
import pandas as pd
import os
import tqdm
import shutil
import random
```

```python
#files_source=glob.glob(r"C:\Users\mfpen\Pictures\train/*/*.png")
def cambiarRuta_imagenes(files_source):
    '''
        Función para cambiar de carpeta train a validacion las imágenes para el entrenamiento.
    '''
    for i in random.choices(files_source,k=200):
        if i.__contains__('area_verde'):

            shutil.move(i, r"\area_verde")
        elif i.__contains__('casas'):

            shutil.move(i, r"\casas")
        elif i.__contains__('establecimiento'):

            shutil.move(i, r"\establecimiento")
        elif i.__contains__('multivivienda'):

            shutil.move(i, r"\multivivienda")
        elif i.__contains__('en_construccion'):

            shutil.move(i, r"\en_construccion")
        elif i.__contains__('terreno_baldio'):

            shutil.move(i, r"\terreno_baldio")
    return 0
```

```python
filenames=glob.glob(r"/*/*.png")+glob.glob(r"/*/*.PNG")
dirs=[]
filename=[]
for file in filenames:
    path=file.replace("\\","/").split("/")
    path[-3]="validacion_pad"
    dirs.append("/".join(path[:-1]))
    filename.append(path[-1])
imagenes=pd.DataFrame({"path":filenames,"dir":dirs,"filename":filename})
imagenes
```

```python
def add_padding(imagenes):
    for carpeta in imagenes["dir"].unique():
        try:
            os.mkdir(carpeta)
            
            for i in tqdm.tqdm(range(len(imagenes))):
                img = cv2.imread(imagenes["path"][i])

                
                old_image_height, old_image_width, channels = img.shape

                # create new image of desired size and color (blue) for padding
                new_image_width = 224
                new_image_height = 224
                color = (0,0,0)
                result = np.full((new_image_height,new_image_width, channels), color, dtype=np.uint8)

                # compute center offset
                x_center = (new_image_width - old_image_width) // 2
                y_center = (new_image_height - old_image_height) // 2

                # copy img image into center of result image
                try:
                    result[y_center:y_center+old_image_height, 
                        x_center:x_center+old_image_width] = img
                    
                except:
                    continue

                cv2.imwrite(imagenes["dir"][i]+"/"+imagenes["filename"][i], result)
        except:
            pass   
    
    return 
```
